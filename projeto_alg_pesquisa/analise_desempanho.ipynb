<h1>Calcular o número de operações T(n) nas funcoes anteriormente criadas:<h1>

<h3>Pesquisa Sequencial<h3>

```python
def pesquisa_sequencial_t(lista, objeto):
    n_operacao = 0
    for i in range(len(lista)):
        n_operacao += 1
        if lista[i] == objeto:
            return i
    return n_operacao,"Não existe"
```

```python
ex_1 = pesquisa_sequencial_t([1,2,3,4,5], 8)
print(ex_1)

ex_1_1 = pesquisa_sequencial_t([1,2,3,4,5], 3)
print(ex_1_1)
```

<h3>Pesquisa Sequencial recursiva<h3>

```python
def pesquisa_sequencial_recursiva_t(lista, n, n_operacao=0):
    if not lista:
        return False, n_operacao
    if lista[0] == n:
        return True, n_operacao + 1
    resultado, n_operacao = pesquisa_sequencial_recursiva_t(lista[1:], n, n_operacao + 1)
    return resultado, n_operacao
```

```python
ex_1_2 = pesquisa_sequencial_recursiva_t([1,2,3,4,5], 3)
print(ex_1_2)
```

<h3>Pesquisa binária<h3>

```python
def pesquisa_binaria_t(lista, item):
    primeiro = 0
    ultimo = len(lista) - 1
    n_operacao = 0 

    while primeiro <= ultimo:
        n_operacao += 1  
        meio = (primeiro + ultimo) // 2
        if lista[meio] == item:
            return True, n_operacao  
        elif item < lista[meio]:
            ultimo = meio - 1
        else:
            primeiro = meio + 1

    return False, n_operacao 
```

```python
ex_1_3 = pesquisa_sequencial_recursiva_t([1,2,3,4,5], 3)
print(ex_1_3)
```

<h3>Pesquisa binária recursiva<h3>

```python
def pesquisa_binaria_recursiva_t(lista, item):
    if not lista:
        return False, 0 

    meio = len(lista) // 2
    n_operacoes = 1 

    if lista[meio] == item:
        return True, n_operacoes  
    elif item < lista[meio]:
        resultado, operacoes_na_sublista = pesquisa_binaria_recursiva_t(lista[:meio], item)
    else:
        resultado, operacoes_na_sublista = pesquisa_binaria_recursiva_t(lista[meio + 1:], item)

    n_operacoes += operacoes_na_sublista
    return resultado, n_operacoes
```

```python
ex_1_4 = pesquisa_sequencial_recursiva_t([1,2,3,4,5], 3)
print(ex_1_4)
```

<h1>Avalie a complexidade temporal de cada um destes casos de forma teórica, identificando o Big-O<h1>

<h3>Pesquisa_sequencial<h3>

<h5>Big-O: O(n), onde n é o número de elementos na lista. Pois no pior caso a pesquisa vai verificar todos os items da lista. Assim quanto maior a lista maior o numero de operacões<h5>

<h3>Pesquisa_sequencial_recursiva<h3>

<h5>Big-O: O(n), onde n é o número de elementos na lista. Pois no pior caso a pesquisa vai procurar todos os items da lista. Assim quanto maior a lista maior o numero de operacões<h5>

<h3>Pesquisa_binaria<h3>

<h5>Big-O: O(log(n)), onde n é o número de elementos na lista. Isso ocorre porque a lista (ordenada) esta a ser susessivamente dividida em dois, assim sendo, o tempo cresce de forma logarítmica em relação ao tamanho da lista.<h5>

<h3>Pesquisa_binaria_recursiva<h3>

<h5>Big-O: O(log(n)), onde n é o número de elementos na lista. Isso ocorre porque a lista (ordenada) esta a ser susessivamente dividida em dois, assim sendo, o tempo cresce de forma logarítmica em relação ao tamanho da lista.<h5>
